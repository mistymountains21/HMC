```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
file = "HMCfinal.csv"
```

```python
df = pd.read_csv(file)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Years to Grad'], df['Entry Age'], alpha=0.7)

# Labeling the axes
plt.xlabel('Time to Graduate (years)')
plt.ylabel('Entry Age')

# Title of the plot
plt.title('Time to Graduate vs. Entry Age')

# Show grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Entry Age', y='Years to Grad', data=df)

# Labeling the axes
plt.xlabel('Entry Age')
plt.ylabel('Time to Graduate (years)')

# Title of the plot
plt.title('Boxplot of Time to Graduate by Entry Age')

plt.xticks(fontsize=10)
plt.savefig('boxplot_entry_age_time_to_graduate.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
X_before = df[['Credits Entry']]
X_after = df[['Credits @ HMC']]
Y = df['Years to Grad']
```

```python
model_before = LinearRegression()
model_before.fit(X_before, Y)

# Create and fit the linear regression model for credits_after
model_after = LinearRegression()
model_after.fit(X_after, Y)

# Print the coefficients and intercepts
print(f'Model (Credits Before): Coefficient: {model_before.coef_[0]}, Intercept: {model_before.intercept_}')
print(f'Model (Credits After): Coefficient: {model_after.coef_[0]}, Intercept: {model_after.intercept_}')
```

```python
plt.figure(figsize=(14, 7))

# Plot for credits_before vs time_to_grad
plt.subplot(1, 2, 1)
sns.scatterplot(x=df['Credits Entry'], y=df['Years to Grad'])
plt.plot(df['Credits Entry'], model_before.predict(X_before), color='red')
plt.title('Credits Before vs Time to Graduate')
plt.xlabel('Credits Before')
plt.ylabel('Time to Graduate')

# Plot for credits_after vs time_to_grad
plt.subplot(1, 2, 2)
sns.scatterplot(x=df['Credits @ HMC'], y=df['Years to Grad'])
plt.plot(df['Credits @ HMC'], model_after.predict(X_after), color='red')
plt.title('Credits After vs Time to Graduate')
plt.xlabel('Credits After')
plt.ylabel('Time to Graduate')

plt.tight_layout()
plt.savefig('linegraph.png')
plt.show()
```
